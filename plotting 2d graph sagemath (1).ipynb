```python
# Piecewise functions
# Ex. Plot the graph of the function f(x) which is 1+x^3 in [0,1] and 1+x in [-1,0].
f=piecewise([((0,1),1+x^3),([-1,0],1+x)]);
show(f)
f.plot(-1,1)
```

```python
#Ex. f(x)={{-1, -1<x<0.5}{1-x^2, 0.5<x<1.5}{2x^3-3,1.5<x<2.5}}
f1(x)=-1
f2(x)=1-x^2
f3(x)=2*x^3-3
f=piecewise([[(-1,0.5),f1],[(0.5,1.5),f2],[(1.5,2.5),f3]])
show(f)
plot(f,(-1,2),figsize=4)
```

```python
# Parametric plot in 2D
# Ex Plot the graph of the function given by parametric coordinates x=tsin(1-t^2), y=tcos(t^2), -2pi<t<2pi
var('x,y,t')
show(x==t*sin(1-t^2))
show(y==t*cos(t^2))
show([-2*pi < t], [t < 2*pi] )
p=parametric_plot([t*sin(1-t^2),t*cos(t^2)],(t,-2*pi,2*pi),color='red',thickness=1.2)
txt=text('Nice plot',(-4,4))
show(p+txt,figsize=5)
```

```python
# Polar Plot
# Ex. Plot the graph of the function given by polar curve r = 1 + 2 cos(sθ), 0 ≤ θ ≤ 200π for various values of s = 1, 1.01, 1.02, 1.05, 1.5 etc
var('theta')
s =2
polar_plot(1+2*cos(s*theta),(theta,0,200*pi),plot_points=5000,thickness=0.5,figsize=8)
```
