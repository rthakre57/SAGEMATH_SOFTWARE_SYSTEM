```python
# Find the 1st four derivative of f(t) = ln(1 + t2) and plot them along with the graph of f(t).

var('t')
f(t) = ln(1+t^2)
show(f)
fp = plot(f(t),(t,-2,2),color='red',legend_label=r'$f$')
f1p = plot(f.diff(t),(t,-2,2),color='green',legend_label=r'$f^{(1)}$')
f2p = plot(f.diff(t,2),(t,-2,2),color='blue',legend_label=r'$f^{(2)}$')
f3p = plot(f.diff(t,3),(t,-2,2),color='pink',legend_label=r'$f^{(3)}$');
f4p = plot(f.diff(t,4),(t,-2,2),color='black',legend_label=r'$f^{(4)}$')
show(fp+f1p+f2p+f3p+f4p,figsize=5)
```

```python
# ii)The derivative dy/dx is obtained by executing the following code in SageMath.
y=function('y')(x)
f = cos(x - sin(y)) - sin(y - sin(x))
show(f)
f1 = diff(f, x)
show(f1)
ans = solve(f1, diff(y))[0].right()
show('dy/dx = ', ans)
```

```python
# The derivative d2y/dx2 is obtained by executing the following code subsequently.
df1 = solve(f1, diff(y))[0].right()
show(df1)
f2 = diff(df1)
show(f2)
ans2 = solve(f2, diff(y))[0].right()
show('dy2/dx2 = ', ans2)

```
