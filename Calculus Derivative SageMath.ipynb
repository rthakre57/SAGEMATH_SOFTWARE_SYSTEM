```python
f(x,y)= y * (y^2-1) * (y-2) - x * (x-1) * (x-2)
dyx = f.implicit_derivative(y,x)
nr = dyx.numerator()
xs = solve(nr==0,x,solution_dict=True)
a, b = xs[0][x],xs[1][x]
a,b
```

```python
y1 = f.subs(x=a)(x,y)
show(y1)
r1 = y1.roots()
show(r1)
for i in range(len(r1)):
    show(r1[i])
```

```python
y2 = f.subs(x=b)(x,y)
r2 = y2.roots()
r2
show(r2)
```

```python
for i in range(len(r2)):
    show(r2[i])
```
